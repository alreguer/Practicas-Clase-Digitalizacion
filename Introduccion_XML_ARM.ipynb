## XML Copy Editor

XML Copy Editor es un programa que permite editar ficheros XML. Sus características, que se pueden consultar en este [enlace](https://xml-copy-editor.sourceforge.io/), son las siguientes:

* Gratuito
* Multiplataforma (Window, Mac, Linux)
* Permite editar y validar documentos XML
* Definir esquemas DTD y XML Schema
* Ejecutar consultas con XPath


### Instalar el programa

Se puede descargar desde este [enlace](https://sourceforge.net/projects/xml-copy-editor/files/).

Elegir la versión que corresponda al sistema operativo disponble (windows, Mac o Linux), descargar el instalador e instalarlo.

La instalación es sencilla. Símplemente sigue los pasos que te marca el instalador.

Una vez instalado, abre el programa.


### Interfaz del programa

Los iconos del programa contemplan las opciones típicas de crear un nuevo fichero, abrir uno existente, guardar o imprimir el fichero.

Además, hay dos iconos de verificación (azul y verde). El primero (azul) sirve para comprabar si un documento está bien formado, mientras que el segundo (verde) para comprobar si un documento XML es válido respecto a un DTD o XML esquema
